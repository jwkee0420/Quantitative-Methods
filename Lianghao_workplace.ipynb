`Casualty is more likely to happen in which age group, and which age group people are more likely to get  more severe injury?`

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import geopandas as gpd
```

```python

selected_col = ['accident_year', 'age_of_casualty', 'casualty_severity']
df_age_analysis = pd.read_csv('data/dft-road-casualty-statistics-casualty-last-5-years.csv', usecols=selected_col)


# df_age_analysis.isna().sum(axis=0) # no na value

df_age_analysis = df_age_analysis[df_age_analysis.age_of_casualty != -1] # ignore missing values
# df_age_analysis['count'] = df_age_analysis['age_of_casualty'].count

grouped = (
  df_age_analysis.groupby(['accident_year', 'age_of_casualty'])
  .agg(
    # count_age=('casualty_severity', 'count'),
    mean_severity=('casualty_severity', 'mean')
  )
  .reset_index()
)
grouped.head(20)
```

```python
# åˆ›å»ºä¸€ä¸ªå›¾è¡¨
plt.figure(figsize=(10,6))
sns.lineplot(x='age_of_casualty', y='mean_severity', hue='accident_year', data=grouped)

# è®¾ç½®æ ‡ç­¾
plt.title("Mean Severity of Casualties by Age", fontsize=16)
plt.xlabel("Age of Casualty", fontsize=12)
plt.ylabel("Mean Severity", fontsize=12)

# æ˜¾ç¤ºå›¾è¡¨
plt.show()
```

```python
# çº¿æ€§å›žå½’æ¨¡åž‹
X = grouped[['age_of_casualty']]  # é¢„æµ‹å˜é‡
y = grouped['mean_severity']  # ç›®æ ‡å˜é‡

model = LinearRegression()
model.fit(X, y)

# é¢„æµ‹å€¼
y_pred = model.predict(X)

# ç»˜åˆ¶æ•£ç‚¹å›¾å’Œå›žå½’çº¿
plt.figure(figsize=(10,6))
sns.scatterplot(x='age_of_casualty', y='mean_severity', data=grouped, color='blue')
plt.plot(grouped['age_of_casualty'], y_pred, color='red', linewidth=2)
plt.title("Linear Regression: Age of Casualty vs Mean Severity", fontsize=16)
plt.xlabel("Age of Casualty", fontsize=12)
plt.ylabel("Mean Severity", fontsize=12)
plt.show()

# æ‰“å°å›žå½’æ¨¡åž‹çš„ç³»æ•°å’Œè¯„ä¼°æŒ‡æ ‡
print(f"Regression Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")
print(f"Mean Squared Error: {mean_squared_error(y, y_pred)}")
print(f"R-squared: {r2_score(y, y_pred)}")
```

```python
# è®¡ç®—çš®å°”é€Šç›¸å…³ç³»æ•°
correlation = grouped['age_of_casualty'].corr(grouped['mean_severity'])
print(f"Pearson Correlation Coefficient: {correlation}")
```

## clustering

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.cluster as sklc
import sklearn.metrics as sklm
```

```python
num_clusters = 5
data_filename = 'data/dft-road-casualty-statistics-collision-2023.csv'
df = pd.read_csv(data_filename)
data = np.genfromtxt(data_filename, delimiter=',')
df.info()
```

## Important!

> If you want to cluster  different columns, you just need to change the **selected_data** variable in the code chunk below, and then run all.  
> But the result looks less meaningful, I think the reason is the severity level is so concentrated, just 3 interal level.

```python
valid_rows =  ~np.isnan(data[:, 12])
data[valid_rows,8] #Accident_Severity
data[valid_rows,20] #Speed_limit
data[valid_rows,28] #Weather_Conditions

#here! 
selected_data = data[valid_rows][:,[20, 8]]

figure_width, figure_height = 7, 7
fig_title = 'Figure Title'
x_lable = 'x-axis label'
y_lable = 'y-axis label'
title_fontsize = 15
label_fontsize = 10

# These lines set the limits of the x and y axes, so that all plots are on the same scale.
x_min, x_max = 0.5*np.min(selected_data[:,0]), 1.1*np.max(selected_data[:,0])
y_min, y_max = 0.5*np.min(selected_data[:,1]), 1.1*np.max(selected_data[:,1])

def setup_figure():
  plt.xlim([x_min, x_max])
  plt.ylim([y_min, y_max])
  plt.xlabel(x_lable, fontsize=label_fontsize)
  plt.ylabel(y_lable, fontsize=label_fontsize)

x_values = selected_data[:, 0]
y_values = selected_data[:, 1]

plt.figure(0, figsize=(figure_width, figure_height))
setup_figure()

plt.xlabel(x_lable, fontsize=label_fontsize)
plt.ylabel(y_lable, fontsize=label_fontsize)
plt.title(fig_title,fontsize=title_fontsize)
plt.plot(x_values, y_values, 'k.')
```

### K-means clustering

```python
kmeans_output = sklc.KMeans(n_clusters=num_clusters, n_init=1).fit(selected_data)

# This line creates a list giving the final cluster number of each point:
clustering_ids_kmeans = kmeans_output.labels_

#Which we can now look at
print(clustering_ids_kmeans)  
```

```python
selected_data
```

```python
#That's just an ordered list of labels, which could be more helpful if they were attached to the data points. 

# These lines add the cluster IDs to the original data and save the data with these added cluster IDs.
complete_data_with_clusters = np.hstack((selected_data,np.array([clustering_ids_kmeans]).T))
'''
â€¢  np.array([clustering_ids_kmeans]).T:
â€¢	é¦–å…ˆï¼Œ[clustering_ids_kmeans] å°† clustering_ids_kmeans è½¬æ¢ä¸ºä¸€ä¸ªåµŒå¥—çš„äºŒç»´åˆ—è¡¨ï¼Œå½¢çŠ¶ä¸º (1Ã—n)(1 \times n)(1Ã—n)ã€‚
â€¢	.T æ˜¯è½¬ç½®æ“ä½œï¼Œå°†å…¶è½¬æ¢ä¸º (nÃ—1)(n \times 1)(nÃ—1)ï¼Œå³æ¯ä¸ªèšç±»æ ‡ç­¾ä½œä¸ºä¸€è¡Œï¼Œå½¢æˆä¸€åˆ—çš„äºŒç»´æ•°ç»„ã€‚
â€¢	è¿™ä¸€æ­¥æ˜¯ä¸ºäº†ç¡®ä¿æ ‡ç­¾å¯ä»¥ä¸Ž data æŒ‰åˆ—æ‹¼æŽ¥ï¼ˆä¸¤è€…çš„è¡Œæ•°ä¸€è‡´ï¼‰ã€‚

â€¢  np.hstack((data, ...)):
â€¢	np.hstack æ˜¯ NumPy çš„æ¨ªå‘æ‹¼æŽ¥å‡½æ•°ï¼Œç”¨äºŽå°†å¤šä¸ªæ•°ç»„æŒ‰åˆ—è¿žæŽ¥ã€‚
â€¢	å®ƒå°†åŽŸå§‹æ•°æ® data å’Œè½¬ç½®åŽçš„ clustering_ids_kmeans æŒ‰åˆ—æ‹¼æŽ¥ï¼Œå½¢æˆä¸€ä¸ªæ–°çš„äºŒç»´æ•°ç»„ã€‚
â€¢	å¦‚æžœ data çš„å½¢çŠ¶æ˜¯ (nÃ—m)(n \times m)(nÃ—m)ï¼Œæ‹¼æŽ¥åŽçš„ complete_data_with_clusters å½¢çŠ¶å°†æ˜¯ (nÃ—(m+1))(n \times (m+1))(nÃ—(m+1))ã€‚
'''

#Have a look at the result:
print(complete_data_with_clusters)
```

```python
# åˆ†ç±»
data_by_cluster = []

for i in range(num_clusters):
    
    this_data = []
    
    for row in complete_data_with_clusters:
        
        if row[-1] == i: #æ¯è¡Œæœ€åŽä¸€åˆ—æ•°æ®ï¼Œå³åˆ†ç±»å€¼
            this_data.append(row)
    
    this_data = np.array(this_data)
    
    data_by_cluster.append(this_data) #åˆ—è¡¨åŒ…array
    
#Which gives the following:
data_by_cluster
```

```python
#Lovely! Now let's have a look at those clusters:
# FIGURES 1-N : SEPARATE CLUSTER PLOTS

# This is a list of colours to differentiate each cluster.
color_list = ['b','r','g','m','c','k','y']

# This loop goes through each cluster, plots it and saves it:
for i in range(num_clusters):
    
    plt.figure(i+1,figsize=(figure_width,figure_height))
    setup_figure()
    plt.title(fig_title + ' - Cluster ' + str(i), fontsize=title_fontsize)
    
    x_values = data_by_cluster[i][:,0]
    y_values = data_by_cluster[i][:,1]
    
    plt.plot(x_values,y_values,color_list[i % num_clusters] + '.')
```

```python
#Might be nice to have those on the same plot though.
# FIGURE N + 1 : COMBINED CLUSTER PLOT

# These lines create a plot with all the data points, coloured by cluster.
plt.figure(num_clusters + 1, figsize=(figure_width,figure_height))
setup_figure()
plt.title(fig_title + ' - Coloured by Cluster',fontsize=title_fontsize)

for i in range(num_clusters):
    
    x_values = data_by_cluster[i][:,0]
    y_values = data_by_cluster[i][:,1]
    
    plt.plot(x_values,y_values,color_list[i % num_clusters] + '.')
```

`Silhouette Analysisï¼ˆè½®å»“åˆ†æžï¼‰æ˜¯ä¸€ç§ç”¨äºŽè¯„ä¼°èšç±»ç»“æžœè´¨é‡çš„æ–¹æ³•ï¼Œé€šå¸¸ç”¨æ¥è¾…åŠ©é€‰æ‹©èšç±»çš„æœ€ä½³ç°‡æ•° kã€‚å®ƒé€šè¿‡è®¡ç®—æ¯ä¸ªæ•°æ®ç‚¹çš„è½®å»“ç³»æ•°ï¼ˆSilhouette Coefficientï¼‰ï¼Œè¡¡é‡è¯¥ç‚¹åœ¨å…¶æ‰€å±žç°‡å†…çš„ç´§å¯†åº¦å’Œä¸Žå…¶ä»–ç°‡çš„åˆ†ç¦»åº¦ã€‚å…¶å–å€¼èŒƒå›´ä¸º [âˆ’1,1]ï¼Œæœ€ä½³ç°‡æ•° k å¯¹åº”çš„å¹³å‡è½®å»“ç³»æ•°åº”è¯¥æŽ¥è¿‘ 1ã€‚`


```python
# SILHOUETTE SCORE

# These lines calculate the silhouette score...
silhouette_kmeans = sklm.silhouette_score(selected_data, clustering_ids_kmeans)

# ... and print it:
print("Silhouette Score:", silhouette_kmeans)
```
